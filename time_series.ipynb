```python
from fbprophet import Prophet
from fbprophet.plot import plot_plotly
import pandas as pd
from datetime import datetime
```

Train data was aquired with yfinance for the last 2 years.

```python
url = 'https://raw.githubusercontent.com/rozumum/eurusd/main/train_data.csv'
df = pd.read_csv(url)
```

```python
df.columns = ['y', 'ds']
```

```python
df.head()
```

TRAIN MODEL

```python
m = Prophet(interval_width=0.95, daily_seasonality=True)
model = m.fit(df)
```

Make a forecast

```python
future = m.make_future_dataframe(periods=30,freq='D')
forecast = m.predict(future)
forecast.head()
```

```python
plot1 = m.plot(forecast)
```

```python
plt2 = m.plot_components(forecast)
```

The prediction of EURUSD for one month.

```python
forecast.tail(30)
```

```python
plot1 = m.plot(forecast[-30:])
```
